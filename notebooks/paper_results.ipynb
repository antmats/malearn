## Results tables

`````python
numerical_imputation = "zero"  # "zero" or "mice"
`````

`````python
table = pd.concat(
    [
        adni_table,
        fico_table,
        life_table,
        nhanes_table,
    ],
    keys=["adni", "fico", "life", "nhanes"],
    names=["dataset"],
)

# Comment out the lines below to generate tables for Breast Cancer and Pharyngitis.
#table = pd.concat(
#    [breast_table, pharyngitis_table],
#    keys=["breast", "pharyngitis"],
#    names=["dataset"],
#)
`````

`````python
# Generate the main table with $\alpha=\alpha^{*}$.

final_table = table[
    table.index.get_level_values("numerical_imputation").isin([numerical_imputation, "none"])
    & (table.index.get_level_values("alpha") == -1)
    & (table.index.get_level_values("estimator_alias") != "mgam_no")
]

final_table = final_table.reindex(DATASET_MAP, level="dataset")
final_table = final_table.reindex(ESTIMATOR_MAP, level="estimator_alias")
final_table = final_table.droplevel("alpha")
final_table = final_table.droplevel("numerical_imputation")

final_table.index = pd.MultiIndex.from_tuples(
    final_table.index.map(
        lambda i: (DATASET_MAP[i[0]], ESTIMATOR_MAP[i[1]])
    ),
    names=["Dataset", "Estimator"],
)

final_table = final_table.rename(columns=METRIC_MAP)

first_dataset = final_table.index.get_level_values(0)[0]
estimator_order = final_table[
    final_table.index.get_level_values(0) == first_dataset
].index.get_level_values(1)

# Choose metrics to present in the table.
metrics = ["roc_auc_ovr", "missingness_reliance"]
metrics = [METRIC_MAP[m] for m in metrics]

final_table = final_table.reset_index().pivot(
    index="Estimator",
    columns=["Dataset"],
    values=metrics,
)

final_table = final_table.reindex(estimator_order)

final_table.columns = final_table.columns.swaplevel(0, 1)
final_table = final_table.sort_index(axis=1, level=0, sort_remaining=False)

final_table = final_table.map(
    lambda x: (
        x.split(" ")[0]
        + " "
        + "{\scriptsize "
        + (x.replace(x.split(" ")[0], "") if not (x.startswith("100.0") or x.startswith("0.0")) else x.removesuffix("100.0").removesuffix("0.0"))
        + " }"
    )
)

print(final_table.to_latex())
`````

`````python
# Compute the average difference in missingness reliance between MA estimators and their unregularized counterparts.

for ma_estimator, estimator in [("MA-Lasso", "LR"), ("MA-DT", "DT"), ("MA-RF", "RF"), ("MA-GBT", "XGB")]:
    rho_columns = [c for c in final_table.columns if c[1] == r"$\hat{\rho}$"]
    
    rho_ma_dt = final_table.loc[ma_estimator, rho_columns]
    rho_dt = final_table.loc[estimator, rho_columns]
    
    rho_ma_dt = rho_ma_dt.apply(lambda x: float(x.split(" ")[0]))
    rho_dt = rho_dt.apply(lambda x: float(x.split(" ")[0]))

    diff = (rho_dt - rho_ma_dt)
    print(f"{ma_estimator} vs. {estimator}: {diff.mean():.1f}")
`````

`````python
# Generate the table with MA estimators using different values for $\alpha$.

final_table = table[
    table.index.get_level_values("numerical_imputation").isin(["zero", "none"])
    & table.index.get_level_values("estimator_alias").isin(["malasso", "madt", "marf", "magbt"])
]

final_table = final_table.reindex(DATASET_MAP, level="dataset")
final_table = final_table.reindex(ESTIMATOR_MAP, level="estimator_alias")
final_table = final_table.droplevel("numerical_imputation")

def rename_index(index):
    d, e, a = index
    if a == -1:
        alpha = r"$\alpha=\alpha^{*}$"
    elif a == 0:
        alpha = r"$\alpha=0$"
    elif a == -2:
        alpha = r"$\alpha=\infty$"
    return (DATASET_MAP[d], f"{ESTIMATOR_MAP[e]} ({alpha})", a)

final_table.index = pd.MultiIndex.from_tuples(
    final_table.index.map(rename_index),
    names=["Dataset", "Estimator", "alpha"],
)

final_table = final_table.droplevel("alpha")
final_table = final_table.rename(columns=METRIC_MAP)

first_dataset = final_table.index.get_level_values(0)[0]
estimator_order = final_table[
    final_table.index.get_level_values(0) == first_dataset
].index.get_level_values(1)

# Choose metrics to present in the table.
metrics = ["roc_auc_ovr", "missingness_reliance"]
metrics = [METRIC_MAP[m] for m in metrics]

final_table = final_table.reset_index().pivot(
    index="Estimator",
    columns=["Dataset"],
    values=metrics,
)

final_table = final_table.reindex(estimator_order)

final_table.columns = final_table.columns.swaplevel(0, 1)
final_table = final_table.sort_index(axis=1, level=0, sort_remaining=False)

final_table = final_table.map(
    lambda x: (
        x.split(" ")[0]
        + " "
        + "{\scriptsize "
        + (x.replace(x.split(" ")[0], "") if not (x.startswith("100.0") or x.startswith("0.0")) else x.removesuffix("100.0").removesuffix("0.0"))
        + " }"
    )
)

print(final_table.to_latex())
`````

## MA estimators: $\alpha=0$ vs. $\rho=0$

`````python
data = (
    adni_scores[
        adni_scores.estimator_alias.str.startswith("ma")
        & adni_scores.alpha.isin([-2, 0])
        & adni_scores.numerical_imputation.eq("zero")
    ]
    .astype(dict(alpha=str))
)

data = data.replace({"estimator_alias": ESTIMATOR_MAP})

estimator_order = [e for e in ESTIMATOR_MAP.values() if e.startswith("MA")]
data["estimator_alias"] = pd.Categorical(data["estimator_alias"], categories=estimator_order, ordered=True)
data.sort_values(by=["estimator_alias", "seed"], ascending=[True, True])
data.estimator_alias = data.estimator_alias.cat.remove_unused_categories()

for y, ylabel, figlabel in [
    ("roc_auc_ovr", "AUROC", "auc"),
    ("missingness_reliance", r"Missingness reliance ($\hat{\rho}$)", "rho")
]:
    fig, ax = plt.subplots(figsize=(6, 4))

    plt.rcParams.update(RC_DEFAULT)

    sns.barplot(
        data=data,
        x="estimator_alias",
        y=y,
        hue="alpha",
        hue_order=["-2.0", "0.0"],
        ax=ax,
    )
    
    ax.xaxis.label.set_color("white")
    ax.set_ylabel(ylabel)
    
    handles, _labels = ax.get_legend_handles_labels()
    labels = [r"$\alpha=\infty$", r"$\alpha=0$"]
    
    ax.legend(handles, labels)
    
    sns.move_legend(ax, "lower left", bbox_to_anchor=(0, 0.1))
    
    fig.savefig(f"life_{figlabel}_vs_estimator.pdf", dpi=300, bbox_inches="tight", pad_inches=0)
`````

## MAR and MNAR missingness

To reproduce this experiment, run:
```bash
./scripts/slurm/run_experiment_wrapper.sh config_cla.yml breast lasso_miss
````

`````python
experiment_dirs = ["../paper_results/250419_1053_breast"]

synthetic_table, synthetic_scores = get_results_table(
    experiment_dirs,
    metrics=METRIC_MAP.keys(),
    groupby=["estimator_alias", "alpha", "numerical_imputation"],
    include_cis=True,
)
`````

`````python
grouped_synthetic_scores = (
    synthetic_scores[
        synthetic_scores.missing_rate.eq(0.5)
        & (
            synthetic_scores.alpha.eq(-1)
            | synthetic_scores.alpha.ge(1)
        )
    ]
    .astype(dict(missing_features=str, missing_rate=str))
    .groupby(["missing_mechanism", "alpha", "missing_features"])
)

mean_synthetic_scores = grouped_synthetic_scores[["missingness_reliance", "roc_auc_ovr"]].mean()

fig, ax = plt.subplots(figsize=(6, 4))

plt.rcParams.update(RC_DEFAULT)

sns.lineplot(
    data=mean_synthetic_scores,
    x="missingness_reliance",
    y="roc_auc_ovr",
    hue="missing_features",
    style="missing_mechanism",
)

ax.grid("on")

from matplotlib.ticker import FormatStrFormatter
ax.yaxis.set_major_formatter(FormatStrFormatter("%.2f"))

ax.set_xlabel(r"Missingness reliance $\left(\hat{\rho}\right)$")
ax.set_ylabel("AUROC")

colors = sns.color_palette()
kwargs = dict(marker="o", s=50, facecolors="w")

ax.scatter(*mean_synthetic_scores.loc[("mar", -1, "0.2")], color=colors[0], linestyle="-", **kwargs)
ax.scatter(*mean_synthetic_scores.loc[("mar", -1, "0.4")], color=colors[1], linestyle="-", **kwargs)
ax.scatter(*mean_synthetic_scores.loc[("mar", -1, "0.6")], color=colors[2], linestyle="-", **kwargs)
ax.scatter(*mean_synthetic_scores.loc[("mnar", -1, "0.2")], color=colors[0], linestyle="--", **kwargs)
ax.scatter(*mean_synthetic_scores.loc[("mnar", -1, "0.4")], color=colors[1], linestyle="--", **kwargs)
ax.scatter(*mean_synthetic_scores.loc[("mnar", -1, "0.6")], color=colors[2], linestyle="--", **kwargs)

ax.legend()

handles, _labels = ax.get_legend_handles_labels()
labels = ["Proportion", "0.2", "0.4", "0.6", "Mechanism", "MAR", "MNAR", "Lasso (MAR)", "Lasso (MNAR)"]

ax.legend(handles, labels)

fig.savefig("breast_missingnes.pdf", dpi=300, bbox_inches="tight", pad_inches=0)
`````

## Comparing training times

`````python
for dataset_alias, experiment_dir in [
    ("adni", "../paper_results/250418_2146_adni"),
    ("fico", "../paper_results/250419_1026_fico"),
    ("life", "../paper_results/250419_1035_life"),
    ("nhanes", "../paper_results/250419_1043_nhanes"),
]:
    for estimator_alias in ["minty", "malasso"]:
        cv_results = collect_cv_results(
            experiment_dir,
            estimator_alias,
            exclude_params={"numerical_imputation": "mice"},
        )
        print(
            DATASET_MAP[dataset_alias],
            ESTIMATOR_MAP[estimator_alias],
            f"{cv_results.mean_fit_time.mean():.1f}"
        )
    print()
`````
